### Limpeza e tratamento de dados

```python
import pandas as pd
import yfinance as yf
```

```python
dados_acao = yf.download(["CASH3.SA", 'VALE3.SA'], "2015-01-01", "2022-01-01")['Adj Close']
dados_acao
```

```python
dados_teste = pd.DataFrame({'empresa': ['Weg', 'Petrobras', 'Vale', pd.NA],
                           'cotacao': [20, 50, pd.NA, pd.NA],
                           'volume': [3000, pd.NA, pd.NA ,pd.NA]})

dados_teste
```

```python
#tirando dados ausentes

dados_acao = dados_acao.dropna()

dados_acao
```

```python
#qualquer NA

print(dados_teste.dropna())

#todas as colunas tem que ser NA
print(dados_teste.dropna(how = 'all'))

#duas colunas como NA

dados_teste.dropna(subset = ["empresa","cotacao"])
```

```python
#prenchendo dados ausentes, troca na pelo oq for pedido

print(dados_teste.fillna(0))

dados_teste.fillna({"empresa":"Nenhuma",
                    "cotacao":0,
                    "volume":-10})
```

```python
cotacoes = pd.Series([20, pd.NA, pd.NA, pd.NA, 23, 24, pd.NA, 25], 
                     index = pd.date_range("2022-06-30", freq = "M", periods = 8))
                                           

cotacoes
```

```python
print(cotacoes.fillna(methotd = 'ffill')) 

cotacoes.fillna(method='ffill',limit = 2)
```

```python
cotacoes.fillna(cotacoes.mean())
```
