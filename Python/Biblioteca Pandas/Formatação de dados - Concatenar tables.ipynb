```python
#essa é uma função que demora para roda. Nunca coloque dentro de um loop

lista_dfs = []

for i in range(10):
    vetor_teste = np.random.random(10)
    teste = ["teste"] * 10

    df = pd.DataFrame({"valores": vetor_teste,
                       "nome":teste})
    
    lista_dfs.append(df)

lista_dfs
```

```python
# Merge x Join - juntar elementos baseados em chaves
#Join é uma versao simplificado do Merge,Merge é aplicado a todos os casos já o Join  a casos especificos do Merge

dados_roe = pd.DataFrame({"ticker":['WEGE3', 'VALE3', 'PCAR3', 'ELET3', 'MGLU3'],
                          "ROE": [20, 9, 19, 3, 5]})

dados_roic = pd.DataFrame({"codigo": ['WEGE3', 'VALE3', 'PCAR3', 'ELET3', 'VIIA3'],
                         "ROIC": [25, 6, 13, 1, 2]})


indicadores = pd.merge(dados_roe,dados_roic,how = "inner",left_on=['ticker'],right_on=['codigo'])

print(dados_roe)
print(dados_roic)

indicadores

```

```python
dados_roe = pd.DataFrame({"ticker": ['WEGE3', 'VALE3', 'PCAR3', 'ELET3'],
                         "ROE": [20, 9, 19, 3]})

dados_roic = pd.DataFrame({"ticker": ['WEGE3', 'VALE3', 'PCAR3', 'ELET3'],
                         "ROIC": [25, 6, 13, 1]})

indicadores = pd.merge(dados_roe, dados_roic, how = "inner", 
                on = "ticker")


indicadores
```

```python
#exercicios 77
from datetime import datetime
from datetime import timedelta

tickers = ["MGLU3.SA","CASH3.SA",'AGRO3.SA']
dia = datetime.now() - timedelta(days = 1)

dados_close= yf.download(tickers, dia)['Close']
dados_close = pd.melt(dados_close, var_name= "cod", value_name= "preco")


dados_volume = yf.download(tickers, dia)['Volume']
dados_volume = pd.melt(dados_volume, var_name= "cod", value_name= "volume")

dados_finais = pd.merge(dados_close, dados_volume, on = "cod")

dados_finais
```
